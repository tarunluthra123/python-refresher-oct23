### Print 1 to 10

````python
# start, end, increment/update
````

````python
i = 1 # initialisation (start)

# condition (end)
while i <= 10:
    print(i)
    i = i + 1 # update (jumps)
````

````python
# Print in the same line
i = 1
while i <= 10:
    print(i, end=', ')
    i = i + 1
````

### Print Odd Numbers from 1 to N

````python
N = int(input())
i = 1

while i <= N:
    print(i, end=' ')
    i = i + 2
````

````python
N = int(input())
i = 1

while i <= N:
    if i % 2 == 1:
        print(i, end=' ')
    i = i + 1
    
# Not an efficient method
# We are visiting all the even numbers as well
# Which is redundant
````

### Given N, print perfect squares of numbers from 1 to N

````python
# N = 5
# 1, 4, 9, 16, 25
````

````python
N = int(input())
i = 1

while i <= N:
    print(i * i, end=' ') # i ** 2
    i += 1
````

### Total sum
Given an integer N as input followed by N integers in the input. Print their sum.

```
3
8 2 4
```

Ans = 14

````python
N = int(input())
i = 1
total = 0

while i <= N:
    x = int(input())
    total += x
    i += 1
    
print("Sum =", total)
````

### Given N, print it's digits (from right to left)

````python
# N = 9473
# Output: 3 7 4 9
````

````python
N = 1234567

while N > 0:
    last_digit = N % 10
    print(last_digit, end=' ')
    N = N // 10
````

# Continue statement
<img src="https://media.giphy.com/media/l2YWxte7sJB2XuE8M/giphy.gif" width="300px" />

Skips an iteration.
Works with both for loops and while loops

````python
i = 1

while i <= 20:
    print(i, end=' ')
    i = i + 1
````

````python
# Skip all the numbers that are divisible by 5
````

````python
i = 0

while i < 20:
    i = i + 1
    if i % 5 != 0:
        print(i, end=' ')
````

````python
i = 0

while i < 20:
    i = i + 1
    
    if i % 5 == 0:
        continue
        
    print(i, end=' ')
````

````python
# Continue statements skips whatever code is left in the current iteration
````
